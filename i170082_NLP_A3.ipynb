**Standard N-Gram Models**

Generating Unigrams and their Frequencies

```python
def get_unigrams_freq(unigrams):
  #counter = Counter(unigrams)
  #return counter
  freq = {}
  for l in unigrams:
    if not l.text in freq:
      freq[l.text] = 1
    else:
      freq[l.text] += 1

  #print(freq)
  return freq
```

Generating Bigrams and their Frequencies

```python
def generate_bigrams(str1):
  bigrams = list()
  for i in range(len(str1)-1):
    w1 = str1[i].text
    w2 = str1[i+1].text
    bigrams.append(tuple((w1, w2)))

  return bigrams
```

```python
def get_bigrams_freq(bigrams):
  freq = {}
  for l in bigrams:
    if not l in freq:
      freq[l] = 1
    else:
      freq[l] += 1

  #print(freq)
  return freq
```

Generating Trigrams and their Frequencies

```python
def generate_trigrams(str1):
  trigrams = list()
  for i in range(len(str1)-2):
    w1 = str1[i].text
    w2 = str1[i+1].text
    w3 = str1[i+2].text
    trigrams.append(tuple((w1, w2, w3)))

  return trigrams
```

```python
def get_trigrams_freq(trigrams):
  freq = {}
  for l in trigrams:
    if not l in freq:
      freq[l] = 1
    else:
      freq[l] += 1

  print(freq)
  return freq
```

Calculating Bigram Probabilities

```python
def calculate_bigram_probability(uni_freq, bi_freq, bigrams):
  prob = {}

  for l in bigrams:
    if not l in prob:
      #print(l[0])
      for b in bi_freq:
        #print((k, bi_freq[k]))
        if (l==b):
          bi_val = bi_freq[b]
      for u in uni_freq:
        if (l[0]==u):
          uni_val = uni_freq[u]

      prob[l] = bi_val/uni_val

  #print(prob)
  return prob
```

Calculating Trigram Probabilities

```python
def calculate_trigram_probability(bi_freq, tri_freq, trigrams):
  prob = {}

  for l in trigrams:
    if not l in prob:
      #print(l[0])
      for t in tri_freq:
        #print((k, bi_freq[k]))
        if (l==t):
          tri_val = tri_freq[t]
      for b in bi_freq:
        if (l[0]==b[0] and l[1]==b[1]):
          bi_val = bi_freq[b]

      prob[l] = tri_val/bi_val

  #print(prob)
  return prob
```

Generating Verses and Stanzas using Bigrams

```python
def generate_verse_using_bigrams(unigrams, bi_prob):
  no_of_words = random.randint(7, 9)
  verse = []

  random_word = random.choice(unigrams)

  bigram = ""
  max_prob = -1

  for j in bi_prob:
    if (j[0]==random_word.text):
      if (bi_prob[j]>max_prob):
        max_prob = bi_prob[j]
        bigram = j
    
  verse.append(bigram[0])
  verse.append(bigram[1])
  #print(verse)

  v = 1

  for i in range(0, no_of_words-1):
    #random_word = random.choice(unigrams)

    bigram = ""
    max_prob = -1

    for j in bi_prob:
      if (j[0]==verse[v]):
        if (bi_prob[j]>max_prob):
          max_prob = bi_prob[j]
          bigram = j
    
    #verse.append(bigram[0])
    verse.append(bigram[1])
    v+=1

  #print(verse)
  verse_str = ""
  for i in range(0, len(verse)):
    verse_str+=(verse[i]+" ")
  verse_str = verse_str[:-1]
  #print(verse_str)
  return verse_str
```

```python
def generate_stanza_using_bigrams(unigrams, bi_prob):
  stanza = []
  for i in range(0, 4):
    stanza.append(generate_verse_using_bigrams(unigrams, bi_prob))

  return stanza
```

```python
def generate_poem_using_bigrams(unigrams, bi_prob):
  poem = []
  for i in range(0, 3):
    stanza = generate_stanza_using_bigrams(unigrams, bi_prob)
    poem.append(stanza)
    for j in range(0, len(stanza)):
      print('،', stanza[j])
    print('\n')

  #print(poem)
  return poem
```

Generating Verses and Stanzas using Trigrams

```python
def generate_verse_using_trigrams(unigrams, bi_prob, tri_prob):
  no_of_words = random.randint(8, 9)
  verse = []

  random_word = random.choice(unigrams)

  bigram = ""
  b_max_prob = -1

  for j in bi_prob:
    if (j[0]==random_word.text):
      if (bi_prob[j]>b_max_prob):
        b_max_prob = bi_prob[j]
        bigram = j

  trigram = ""
  t_max_prob = -1

  for j in tri_prob:
    if (j[0]==bigram[0] and j[1]==bigram[1]):
      if(tri_prob[j]>t_max_prob):
        t_max_prob = tri_prob[j]
        trigram = j
    
  verse.append(trigram[0])
  verse.append(trigram[1])
  #print(verse)

  v = 1

  for i in range(0, no_of_words-5):
    #random_word = random.choice(unigrams)

    bigram = ""
    max_prob = -1

    #print("verse[v]=", verse[v])

    for j in bi_prob:
      if (j[0]==verse[v]):
        if (bi_prob[j]>max_prob):
          max_prob = bi_prob[j]
          bigram = j

    trigram = ""
    t_max_prob = -1

    for j in tri_prob:
      if (j[0]==bigram[0] and j[1]==bigram[1]):
        if(tri_prob[j]>t_max_prob):
          t_max_prob = tri_prob[j]
          trigram = j
    
    #print("T =", trigram)
    #verse.append(bigram[0])
    verse.append(trigram[1])
    verse.append(trigram[2])
    v+=2

  #print(verse)
  verse_str = ""
  for i in range(0, len(verse)):
    verse_str+=(verse[i]+" ")
  verse_str = verse_str[:-1]
  #print(verse_str)
  return verse_str
```

```python
def generate_stanza_using_trigrams(unigrams, bi_prob, tri_prob):
  stanza = []
  for i in range(0, 4):
    stanza.append(generate_verse_using_trigrams(unigrams, bi_prob, tri_prob))

  return stanza
```

```python
def generate_poem_using_trigrams(unigrams, bi_prob, tri_prob):
  poem = []
  for i in range(0, 3):
    stanza = generate_stanza_using_trigrams(unigrams, bi_prob, tri_prob)
    poem.append(stanza)
    for j in range(0, len(stanza)):
      print('،', stanza[j])
    print('\n')

  #print(poem)
  return poem
```
